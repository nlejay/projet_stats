## 4.1 Test du Chi2

```python
# Vérification des valeurs prises par la variable Gender
df_mooc['Gender'].unique()
```

```python
# Recodage des valeurs de la variable Gender.
df_mooc['Gender'] = df_mooc['Gender'].replace({'un homme':'homme', 'une femme':'femme'})
```

Valeurs manquantes des variables *Gender* et *HDI_rec*.

```python
print(df_mooc['Gender'].isna().sum())
print(df_mooc['HDI_rec'].isna().sum())
```

- Tableau de contingence des variables *Gender* et *HDI_rec* (par défaut, la méthode crosstab de pandas ignore les valeurs manquantes).

```python
HDI_vs_gender = pd.crosstab(df_mooc['Gender'],df_mooc['HDI_rec'])
HDI_vs_gender
```

```python
effectif = HDI_vs_gender.sum().sum()
print("Nombre total d'inscrits pris en compte dans l'analyse :", effectif)
```

Nous avons plus de 5 valeurs pour chaque couple (genre, HDI). Les conditions sont réunies pour le test du chi2.

Test d'indépendance du chi2 :
- Hypothèse nulle $H_0$ : *HDI_rec* et *Gender* sont indépendantes.
- Hypothèse alternative $H_1$ : *HDI_rec* et *Gender* ne sont pas indépendantes.
- Seuil de signification : $\alpha$ = 0.05.


```python
import pingouin

expected, observed, result = pingouin.chi2_independence(data=df_mooc, x='Gender', y='HDI_rec')

display("Tableau de contingence : ",observed)
display("Valeurs attendues : ",expected)
display("Résultats du test : ", result.loc[result['test']=='pearson'])
```

**p_value < 0.05 donc on rejette l'hypothèse selon laquelle *HDI_rec* et *Gender* sont indépendantes.**

- Calcul des résidus de Pearson

```python
residuals = pd.DataFrame((observed - expected) / np.sqrt(expected), index=observed.index, columns=observed.columns)
residuals
```

- Calcul du V de Cramer

    $\small V = \sqrt{\dfrac{\chi^2}{n\times min(r-1,c-1)}}$

    où :
    - $\chi^2$ est la statistique du test de $\chi^2$
    - $n$ est l'effectif de l'échantillon
    - $r$ et $c$ sont respectivement le nombre de lignes et de colonnes du tableau de contingence
    
Ici, le minimum de $\small r-1$ et $\small c-1$ est 1. Seul le numérateur est à calculer.

```python
V = np.sqrt(result.loc[result['test']=='pearson','chi2'].values/effectif)
print("V de Cramer : V =", round(V[0],3))
```

## 4.2 Mosaic plot

Pour créer un mosaic plot dont les couleurs suivent les valeurs des résidus, je vais associer à chaque résidu une couleur d'une colormap matplotlib. Mais pour cela, je dois changer l'échelle des résidus pour obtenir des valeurs entre 0 et 1. Pour obtenir une couleur différente pour les résidus négatifs et les résidus positifs, je vais faire en sorte que le changement d'échelle ramène les valeurs négatives dans l'intervalle $\scriptsize[0\,;\,0,5[$ et les valeurs positives dans l'intervalle $\scriptsize [0,5\,;\,1]$.

Détail du calcul :

soit *max* le maximum entre le résidu maximal et la valeur absolue du résidu minimal. En divisant chaque résidu par *max*, on obtient des valeurs entre -1 et 0 pour les résidus négatifs et entre 0 et1 pour les résidus positifs. Pour obtenir des valeurs entre -0.5 et 0.5, il faut donc diviser par 2*max. Il ne reste plus qu'à additionner 0.5 aux résultats pour obtenir des valeurs entre 0 et 0.5 pour les résidus négatifs et entre 0.5 et 1 pour les résidus positifs.

Attention : les valeurs obtenues n'ont pas vocation à être interprétées et ne permettent que de définir une échelle de couleur.

```python
# transformation des valeurs de residus
res_maxi = residuals.max().max()
res_mini = residuals.min().min()
diviseur = 2*max(res_maxi, abs(res_mini))
residuals_norm = pd.DataFrame()

for col in residuals.columns:
    residuals_norm[col] = residuals[col].map(lambda x : x/diviseur+0.5) 
residuals_norm
```

Création du mosaic plot.

```python
import statsmodels.graphics.mosaicplot as sm
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.rcParams['figure.dpi'] = 100

#Utilisation de la police de base de Latex
plt.rc('font', family = 'serif', serif = 'cmr10')
plt.rcParams.update({"text.usetex": True, "axes.formatter.use_mathtext" : True})

#Sélection du dégradé de couleurs pour le graph
cmap = plt.colormaps['coolwarm_r']

#_________________Attribution d'une couleur à chaque valeur de résidu________________________

resid_unstack = residuals.unstack().reset_index()
couleurs_graph = dict()

# pour chaque couple (genre,hdi), attribution dans le dico couleurs_graph de la couleur correspondante de cmap
for i, row in resid_unstack[['Gender','HDI_rec']].iterrows():
    residu = residuals_norm.loc[row['Gender'], row['HDI_rec']]
    couleurs_graph[(row['Gender'], row['HDI_rec'])] = cmap(residu)
    
#fonction d'attribution des couleurs dans le mosaic plot
props = lambda key: {'color': couleurs_graph[key]}

#_______________________Attribution d'un label à chaque rectangle____________________________

# Même procédure d'attribution que pour les couleurs mais avec les effectifs des couples (gender, hdi)
labels_graph = dict()
for i, row in resid_unstack[['Gender','HDI_rec']].iterrows():
    lab = HDI_vs_gender.loc[row['Gender'], row['HDI_rec']]
    labels_graph[(row['Gender'], row['HDI_rec'])] = f'Effectif = {lab}'

labels = lambda k: labels_graph[k]

#_______________________________Création de la figure___________________________________  

fig, ax = plt.subplots(figsize=(7,7))
mosaic = sm.mosaic(df_mooc, index=['Gender', 'HDI_rec'],
                   properties=props,
                   labelizer=labels,
                   gap=0.02, ax=ax)

#text, titre et labels
plt.xlabel('Genre', size=12)
plt.gca().xaxis.set_label_coords(0.5, -0.1)   #placer xlabel en dessous du graphe
plt.text(-0.1, 0.5, 'HDI', rotation=90, va='center', ha='center', size=12)   #ylabel
plt.title("Participants au MOOC suivant le genre et l'HDI (Human Development Index)", y=1.05, size=12)
plt.yticks(fontsize=12)
plt.yticks(fontsize=12)

plt.text(1.3, 0, "HDI :\n - B : Bas\n - I : Intermédiaire\n - TH : Très Haut",
         size=11,
         bbox={'boxstyle':'round', 'color':'whitesmoke'})

p = result.loc[result['test']=='pearson', 'pval']
text_pvalue = "Test d'indépendance du Chi-2 :\n"+r"$p-value<0.0001^{***}$"
plt.text(1.3, 0.15, text_pvalue,
         size=11,
         bbox={'boxstyle':'round', 'color':'whitesmoke'})

#Création d'une échelle de couleur pour la légende
sm = plt.cm.ScalarMappable(cmap=cmap)
sm.set_clim(vmin=-diviseur/2, vmax=diviseur/2)
plt.colorbar(sm, ticks=[-10,-4,-2,0,2,4,10], orientation='vertical', ax=ax).set_label(label=f"Résidus de Pearson", size=12)

#Sauvegarde
image = plt.gcf()
image.savefig("mosaicplot_hdi_genre.png", dpi=200, bbox_inches="tight")
#Affichage
plt.show()
```

**- Parmi les participants au MOOC étudié, les femmes sont sous-représentées dans les pays à HDI bas contrairement aux hommes qui sont surrepérentés.**

# 5. Modèle linéaire, tests non paramétriques
## 5.1 Nombre de vidéos vues par genre

Nous voulons comparer le nombre de vidéos vues par les femmes et par les hommes. Commençons par calculer le nombre moyen de vidéos vues.

```python
df_mooc.groupby('Gender').agg(moyenne_vids=pd.NamedAgg('Nb_vid','mean'),
                              variance=pd.NamedAgg('Nb_vid','var'),
                              effectif=pd.NamedAgg('Nb_vid','count')
                             )
```

Dans notre échantillon, les femmes ont regardé plus de vidéos que les hommes en moyenne. Cependant, un test d'hypothèses est nécessaire pour vérifier si cette différence est significative. Nous allons vérifier si les conditions sont réunies pour effectuer un test de Student. Dans le cas contraire, un test de Mann-Whitney sera effectué. Les échantillons étant indépendants, nous devons vérifier la normalité de leurs distributions et l'homogénéité de leurs variances.

### 5.1.1 Vérification de la normalité de la distribution du nombre de vidéos vues

Observons dans un premier temps la distribution du nombre de vidéos vues. Etant donné la taille des échantillons (largement supérieure à 30), une faible violation de la normalité ne devrait pas compromettre l'utilisation d'un test de Student. En revanche, si la distribution des échantillons est trop éloignée d'une distribution normale, on préfèrera un test non paramétrique de Mann-Whitney.

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(5,5))
sns.histplot(data=df_mooc, x='Nb_vid', hue='Gender', alpha=0.5, palette={'femme':'red', 'homme':'tab:blue'})
plt.title("Distribution du nombre de vidéos vues par les participants au MOOC", fontsize=14)
plt.xlabel("Nombre de vidéos visualisées")
plt.ylabel("Nombre d'apprenants")
plt.legend(title='Genre', loc='upper center', labels=['Hommes', 'Femmes'])
image = plt.gcf()
image.savefig('vid_dist.png', dpi=200, bbox_inches="tight")
plt.show()
```

La distribution n'est pas normale. Vérification supplémentaire par un Q-Q plot.

```python
femmes = df_mooc.loc[~(df_mooc['Gender'].isna()) & (df_mooc['Gender']=='femme')]
hommes = df_mooc.loc[~(df_mooc['Gender'].isna()) & (df_mooc['Gender']=='homme')]

fig, ((ax1, ax2)) = plt.subplots(1,2, figsize=(10,4))
pingouin.qqplot(femmes['Nb_vid'], dist='norm', color='tab:blue', ax=ax1)
ax1.set_title('Femmes')
pingouin.qqplot(hommes['Nb_vid'], dist='norm', color='tab:blue', ax=ax2)
ax2.set_title('Hommes')
plt.suptitle('Comparaison de la distribution du nombre de vidéos vues à une distribution normale')
fig.tight_layout()
image = plt.gcf()
image.savefig('qqplot_nb_vid.png', dpi=200, bbox_inches="tight")
plt.show()
```

Test de normalité de Kolmogorov-Smirnov. Hypothèses pour chaque population (hommes et femmes) :

- Hypothèse nulle $H_0$ : la distribution de l'échantillon suit une loi normale
- Hypothèse alternative $H_1$ : la distribution n'est pas normale
- Seuil de signification $\small \alpha = 0,05$


```python
from scipy.stats import kstest

print(kstest(femmes['Nb_vid'], 'norm'))
print(kstest(hommes['Nb_vid'], 'norm'))
```

Les p-values sont inférieures à 0,05 donc on en conclut que les deux échantillons ne suivent pas un distribution normale.

**Le test de Kolmogorov-Smirnov et la vérification visuelle nous donne la même conclusion : la distribution du nombre de vidéos n'est pas normale et un test de Student n'est pas approprié.**

Test de Levene

- Hypothèse nulle $H_0$ : les variances du nombre de vidéos vues sont égales dans les deux populations (hommes et femmes)

- Seuil de signification $\small \alpha=0.05$


Les degrés de libertés sont :

### 5.2.2 Test de corrélation de Pearson

Ce test est adapté lorsque les distributions des variables sont normales, ce qui n'est pas le cas ici. On nous demande cependant de faire ce test.

- Hypothèse nulle : le coefficient de correlation est égale à 0 (pas de corrélation linéaire)
- Hypothèse alternative : le coefficient est significativement différent de 0 (existence d'une corrélation linéaire).
- Seuil de signification : $\small\alpha = 0,05$

```python
pingouin.corr(quizz_vs_vids['Nb_vid'], quizz_vs_vids['Nb_quizz'], alternative='two-sided', method='pearson')
```

**Le coefficient de correlation est de 0,82 et la p-value est inférieure à 0,05. Si les variables étaient adaptées à un test de corrélation de Pearson, on en déduirait que la corrélation linéaire est significative et qu'elle est élevée.**

### 5.2.3 Test de corrélation de Spearman

Ce test ne nécessite pas que les variables suivent une loi normale. Il vérifie s'il y a une corrélation monotone entre les deux variables (même si cette corrélation n'est pas linéaire).

- Hypothèse nulle : le coefficient de correlation est égale à 0 (pas de corrélation monotone)
- Hypothèse alternative : le coefficient est significativement différent de 0 (existence d'une corrélation monotone).
- Seuil de signification : $\small\alpha = 0,05$

```python
pingouin.corr(quizz_vs_vids['Nb_vid'], quizz_vs_vids['Nb_quizz'], alternative='two-sided', method='spearman')
```

**On obtient une p-value inférieure à 0,05 donc la corrélation est significative. Le coefficient de corrélation est 0,81 donc la corrélation est forte.**

### 5.3.2 ANOVA sans interaction entre les variables indépendantes

Pour vérifier s'il y a un effet significatif sur le nombre de vidéos, on peut effectuer une ANOVA. Cependant, il faut vérifier que nous sommes dans des conditions de validité de l'ANOVA.

```python
#Définition d'une fonction d'affichage de graphiques de diagnostics d'un modèle

def diagnostic_plots(fitted, resid, norm_resid, leverage, alpha=0.5):
    """Fonction qui affiche les graphiques de diagnostic d'un modèle:
    - residus vs valeurs prédites
    - qqplot résidus vs distribution normale
    - scale location
    - résidus vs effet de levier
    
    paramètres :
    - fitted : valeurs prédites
    - resid : résidus
    - norm_resid : résidus normalisés
    - leverage : effet de levier
    - alpha : opacité des points (défaut=0.5)
    """
    
    #racine carrée des résidus standardisés
    resid_sqrt = np.sqrt(np.abs(norm_resid))
    
    # Graphiques de diagnostic d'un
    fig,((ax00, ax01), (ax10,ax11)) = plt.subplots(2,2,figsize=(7,7))

    sns.regplot(x=fitted, y=resid, lowess=True, line_kws={'color':'red'}, scatter_kws={'alpha':alpha}, ax=ax00)
    ax00.set_title('Résidus en fonction des valeurs prédites', size=10)
    ax00.set_xlabel('Valeurs prédites')
    ax00.set_ylabel('Résidus')

    pingouin.qqplot(resid, dist='norm', color='tab:blue', ax=ax01)
    ax01.set_title('Q-Q plot de comparaison à une distribution normale', size=10)
    ax01.set_xlabel('Quantiles de la distribution normale')
    ax01.set_ylabel('Quantiles des résidus')

    sns.regplot(x=fitted, y=resid_sqrt, lowess=True, line_kws={'color':'red'}, scatter_kws={'alpha':alpha}, ax=ax10)
    ax10.set_title('Scale-location', size=10)
    ax10.set_xlabel('Valeurs prédites')
    ax10.set_ylabel('Racine carrée des résidus standardisés')

    sns.regplot(x=leverage, y=resid_sqrt, lowess=True, line_kws={'color':'red'}, scatter_kws={'alpha':alpha}, ax=ax11)
    ax11.set_title('Résidus vs effet de levier', size=10)
    ax11.set_xlabel('Effet de levier')
    ax11.set_ylabel('Résidus standardisés')

    plt.tight_layout()
    image=plt.gcf()
    image.savefig("diagnostic_anova.png", dpi=200, bbox_inches="tight")
    plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#Création modèle linéaire
model_no_interact = ols('Nb_vid ~ C(HDI_rec) + C(Gender)', data=nbVid_vs_hdiGenre).fit()


# Valeurs prédites
fitted_lm = model_no_interact.fittedvalues
# Résidus
residus_lm = model_no_interact.resid
#residus standardisés et leur racine carrée
influence_lm = model_no_interact.get_influence()
residus_norm_lm = influence_lm.resid_studentized_internal
#effet de levier
leverage_lm = influence_lm.hat_matrix_diag

# Graphiques de diagnostic
diagnostic_plots(fitted_lm, residus_lm, residus_norm_lm, leverage_lm)
```

```python
leverage_lm
```

En se concentrant sur les deux premiers graphiques, on peut noter :
- l'homoscédasticité semble respectée (la ligne de tendance du graphique résidus vs valeurs prédites reste autour de 0, la variabilité des résidus semble semblable quelle que soit la valeur de chaque valeur prédite)
- la distribution des résidus ne suit pas du tout une loi normale. En particulier, pour les petites valeurs, les résidus sont plus grands que ce qui est attendu pour une distribution normale. 

Cela remet en cause la validité de l'ANOVA. Cependant, pour l'exercice nous ferons comme si les conditions d'application d'une ANOVA étaient réunies.

ANOVA :

- Hypothèse nulle $H_0$ pour chaque variable indépendante : le nombre moyen de vidéos vues est le même dans chaque groupe
- Hypothèse alternative $H_1$ : le nombre moyen de vidéos vues est différents suivant les groupes
- Seuil de signification $\small \alpha = 0,05$

```python
anova_table = sm.stats.anova_lm(model_no_interact, typ=2)
anova_table
```

- **Les résultats de cette ANOVA montrent une différence très significative concernant le nombre moyen de vidéos vues suivant l'HDI du pays des apprenants ($\scriptsize F(2,8953)=283.55$ ; $\scriptsize p =  4\times 10^{-120}$)**
- **En revanche, on ne peut pas rejeter l'hypothèse que le nombre moyen de vidéos vues est le même pour les hommes et les femmes ($\scriptsize F(1,8953)=0.49$ ; $\scriptsize p =  0.49$)**

```python
display(model_no_interact.summary())
```

Un HDI intermédiaire ou très haut apporte un effet significativement positif sur le nombre moyen de vidéos vues par rapport à un HDI bas (dans chaque cas, p-value<0.05). En revanche, il n'y a pas d'effet significatif le fait d'être un homme comparé à une femme (t = -0.7, p=0.5)

**Test post-hoc de Tukey**

Déterminons entre quels HDI le nombre moyen de vidéos vues est significativement différent. 

```python
import statsmodels.stats.multicomp as multi

mc_HDI = multi.MultiComparison(nbVid_vs_hdiGenre['Nb_vid'], nbVid_vs_hdiGenre['HDI_rec'])

# Effectuer le test post-hoc de Tukey
result = mc_HDI.tukeyhsd()

# Afficher les résultats
print(result)
```

**Les différents résultats ci-dessus montre une différence significative pour chaque paire de HDI (p<0.05).**

### 5.3.3 ANOVA avec interaction entre les variables *Gender* et *HDI_rec*

Là encore, on commence par la création du modèle linéaire et la vérification de la validité d'une ANOVA.

```python
#Modèle linéaire
model_interact = ols('Nb_vid ~ C(HDI_rec) + C(Gender) + C(Gender):C(HDI_rec)', data=nbVid_vs_hdiGenre).fit()

# Valeurs prédites
fitted_inter = model_interact.fittedvalues
# Résidus
residus_inter = model_interact.resid
#residus standardisés et leur racine carrée
influence_inter = model_interact.get_influence()
residus_norm_inter = influence_inter.resid_studentized_internal
#effet de levier
leverage_inter = influence_inter.hat_matrix_diag

# Graphiques de diagnostic
diagnostic_plots(fitted_inter, residus_inter, residus_norm_inter, leverage_inter)
```

Comme pour le modèle sans interaction, les résidus ne suivent pas une distribution normale et les conditions ne sont pas réussies pour effectuer une ANOVA. Mais pour l'exercice, nous la ferons tout de même.

ANOVA :

- Hypothèse nulle $H_0$ pour chaque variable indépendante et la relation entre les variables : le nombre moyen de vidéos vues est le même dans chaque groupe
- Hypothèse alternative $H_1$ : le nombre moyen de vidéos vues est différents suivant les groupes
- Seuil de signification $\small \alpha = 0,05$

```python
anova_table = sm.stats.anova_lm(model_interact, typ=2)
anova_table
```

- Là-encore, on peut dire qu'il y a une différence significative du nombre moyen de vidéos vues suivant l'HDI du pays des apprenants : $\scriptsize F(2,8951)= 283.6$ ; $\scriptsize p=4\times 10^{-20}$
- En revanche on le genre et l'interaction HDI-genre n'ont pas une influence significative : $\scriptsize F(1,8951)= 0.5$ ; $\scriptsize p=0.5$ pour le genre et $\scriptsize F(2,8951)= 1.5$ ; $\scriptsize p=0.2$ pour l'interaction HDI-genre.

```python
display(model_interact.summary())
```

Là-encore on ne peut conclure qu'un HDI intermédiaire ou très haut à un effet significativement positif sur le nombre de vidéos vues par rapport à celui d'un HDI bas. Il n'y a pas d'effet significatif du genre et de l'interaction entre genre et HDI, les p-values sont supérieures à 0.05.

## 6.2 Données de comtage et loi de Poisson

Nous allons revenir sur le nombre de vidéos vues, cette fois en établissant un modèle de régression de Poisson. En effet, la variable sur le nombre de vidéos vues contient des données de comptage. 

Une variable dont la distribution suit une loi de Poisson de paramètre $\lambda$ a une moyenne et une variance égales à $\lambda$. Regardons si la moyenne et la variance du nombre de vidéos vues sont égales.

```python
print("Moyenne :", np.mean(nbVid_vs_hdiGenre['Nb_vid']))
print("Variance :", np.var(nbVid_vs_hdiGenre['Nb_vid']))
```

La variance est 10 fois supérieure à la moyenne. On est dans un cas de surdispersion. Rappelons la distribution du nombre de vidéos vues.

```python
sns.histplot(data=df_mooc, x='Nb_vid', bins=30, color='tab:blue')
plt.title("Distribution du nombre de vidéos vues par les participants au MOOC", fontsize=14)
plt.xlabel("Nombre de vidéos visualisées")
plt.ylabel("Nombre d'apprenants")
plt.show()
```

On voit en effet que cette distribution ne correspond pas tout à fait à une loi de Poisson. On note en particulier un excès de valeurs extrêmes. Appliquons une régression de Poisson et calculons le paramètre de surdispersion.

```python
poisson_model = glm("Nb_vid ~ Gender + HDI_rec", data= nbVid_vs_hdiGenre, family = sm.families.Poisson()).fit()
poisson_model.pearson_chi2/poisson_model.df_resid
```

Ce résultat, très supérieur à 1, confirme la surdispersion. Ici, voici deux explications sur le fait que notre variable ne suit pas réellement une loi de Poisson.

- Les événements ne sont pas indépendants. En effet, le MOOC est sans-doute progressif et un apprenant qui n'a pas visualisé une vidéo n°i a moins de chances de visualiser les suivantes qu'une personne ayant vu cette $\text{i}^\text{ème}$ vidéo.
- Le nombre de vidéos est limité à 30. Ainsi, tous les apprenants ayant plus de 29 vidéos ont vu 30 vidéos exactement au lieu de se répartir sur des valeurs comme 30, 31, 32,... Cela provoque un pic dans la distribution au niveau de la valeur 30.
